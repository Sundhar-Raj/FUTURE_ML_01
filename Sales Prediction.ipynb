```python
df['Description']=df['Description'].fillna('NULL')
df['CustomerID']=df['CustomerID'].fillna('unknown')
```

```python
df=df.drop_duplicates()
```

```python
df['sales']=df['Quantity']*df['UnitPrice']
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])

df['Date']=df['InvoiceDate'].dt.date
df['Time']=df['InvoiceDate'].dt.time
```

```python
df['year']=df['InvoiceDate'].dt.year
df['month']=df['InvoiceDate'].dt.month
df['day']=df['InvoiceDate'].dt.day
df['dayofweek']=df['InvoiceDate'].dt.dayofweek
df['is_weekend']=df['dayofweek'].isin([5,6]).astype(int)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df['sales'],lags=30)
```

An ACF plot is plotted to identify any seasonality within a month

```python
df['lag_1']=df['sales'].shift(1)
df['lag_7']=df['sales'].shift(7)
df['lag_30']=df['sales'].shift(30)
df['lag_365']=df['sales'].shift(365)
df = df.dropna().reset_index(drop=True)
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

df['tokens']=df['Description'].apply(lambda x:word_tokenize(str(x).lower()))
```

we use Word2Vec to convert the description into word embeddings

```python
w2v_model=Word2Vec(sentences=df['tokens'],vector_size=100,window=5,min_count=1,workers=4)
```

```python
import numpy as np

def sentence_embedding(tokens):
    vectors=[]
    for word in tokens:
        if word in w2v_model.wv:
            vectors.append(w2v_model.wv[word])
    if len(vectors)>0:
        return np.mean(vectors,axis=0)
    else:
        return np.zeros(100)

df['embedding']=df['tokens'].apply(sentence_embedding)
            
```

```python
X_embedded=np.vstack(df['embedding'].values)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=0.90)
X_embedded=pca.fit_transform(X_embedded)
```

```python
X_embedded.shape
```

Dimensionality Reduction has been done using PCA

```python
freq_countries=df['Country'].value_counts().nlargest(5).index
df['Country_Grouped']=df['Country'].where(df['Country'].isin(freq_countries),'Other')
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(df[['Country_Grouped']])
```

```python
encoded_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(['Country_Grouped']))
df=pd.concat([df,encoded_df],axis=1)
```

```python
print([col for col in df.columns if col.startswith('Country_Grouped')])
```

Categorical variable country has been encoded using One-hot encoder

```python
df['Total_purchases']=df.groupby('CustomerID')['CustomerID'].transform('count')
```

A new column "Total_purchase" has been included that holds the count of number purchases relative to Custome ID, transformation has been done so that the number of rows of this column is same as the other in the df

```python
other_features=df[['lag_1','lag_7','lag_30','lag_365','Quantity','UnitPrice','Total_purchases','Country_Grouped_EIRE','Country_Grouped_France','Country_Grouped_Germany','Country_Grouped_Other','Country_Grouped_Spain','Country_Grouped_United Kingdom','month','year','dayofweek','is_weekend']].values
```

```python
X=np.hstack([X_embedded,other_features])
```

```python
print(X_embedded.shape)
print(other_features.shape)
```

```python
from sklearn.model_selection import train_test_split

Y=df['sales']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=41)
```

```python
from xgboost import XGBRegressor

model=XGBRegressor(n_estimators=300,max_depth=6,learning_rate=0.05,random_state=42)

model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score
mse=mean_squared_error(Y_test,y_pred)
print('MSE:',mse)
print('RMSE:',np.sqrt(mse))
```

```python
from sklearn.metrics import r2_score
print("R2:", r2_score(Y_test, y_pred))
```

```python
baseline_rmse = np.sqrt(mean_squared_error(Y_test,np.full_like(Y_test, Y_train.mean())))

model_rmse = np.sqrt(mean_squared_error(Y_test, y_pred))

print("Baseline RMSE:", baseline_rmse)
print("Model RMSE:", model_rmse)
```
